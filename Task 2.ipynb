### 1: Preprocess twitter data using Spacy, regex.

* All the package that we need. stop_words file contains the list of word need to remove

````python
import math

import spacy
from spacy.lang.en import English


import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import re

from stop_words import STOP_WORDS
````

* Load the English package of spacy and also load data from csv file

````python
nlp = spacy.load('en')
file = pd.read_csv('Data.csv', encoding='latin1')
data = file['Text'][0:10000]
````

* I use regex to remove punctuations, icons, unnecessary symbols, redundant characters and combine with the STOP_WORDS list to remove common words
* The regex remove_list2 will turn word like ohhhhhhh => oh, then oh will be remove because it in the stop word list

````python
def remove_stop_word(data):
    l = len(data)
    remove_list = re.compile('(:(?!\/\/))|(\.$)|(\.(?=\s+))|(\.(\.+))|([^a-zA-Z0-9@ :\'\.\/+=?-_])|(\'s|\'S)|(\s[+])|(\'RE|\'Re|\'re)') #remove icons and adundant characters
    remove_list2 = re.compile(r'([a-zA-Z])\1\1+')
    for index in range(0,l):
        data.loc[index] = re.sub(remove_list, '', data.loc[index])
        data.loc[index] = re.sub(remove_list2, r'\1', data.loc[index])
        data.loc[index] = ' '.join([word for word in data.loc[index].split() if word.lower() not in STOP_WORDS])
    return data
    
data = remove_stop_word(data)
````

### 2: Extract token's type.

* Create new class for handling the output results, the 'values' attribute stores the sentences in the abbreviation from(Ex: Peter is 20 year old => NAME-ENT NUM year old). The 'types' attribute stores sentences in the type fomat(use for analysing). And 'tokens' store the tokens (in original form) that we get after tokenizing using Spacy, from the 'tokens' attribute, we can get the word-vector of each word and lots of important attribute of word. I store all the information in the pandas series so that we can handle it easily and can save to csv file for checking and storing purpose.

````python
class Token(object):
    def __init__(self, length=1):
        self.values = pd.Series(data=([] for index in range(0,length)))
        self.types = pd.Series(data=([] for index in range(0,length)))
        self.tokens = pd.Series(data=([] for index in range(0,length)))
````

* This is a lemmatization process, the function takes the dataframe that stores all the tweeet (read from csv file) as input and gives back Token object as output.
* We will process each series (each tweets) in the dataframe, tokenize it then check each token that if it is suitable to keep. Here I use regex to match the tagname and use some spacy attribute to check also.

````python
#Lemmatization
def Lemma(data):
    '''
    data: Dataframe that store all the tokens.
    This function return an object and also remove some adundant words leave in the token series
    '''
    length = len(data)
    tokenObject = Token(length) #Create object
    name_tag = re.compile('(@([a-zA-Z0-9_-]+))(?=(\s+|$))')
    for index in range(0, length):
        #Tokenize
        token = nlp(data.loc[index])
        for index2 in range(0, len(token)):
            t = token[index2]
            if t.lemma_ not in STOP_WORDS:
                if(re.match(name_tag, t.text)):                   
                    tokenObject.values.loc[index].append('TAGNAME')
                    tokenObject.types.loc[index].append('TAGNAME')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.like_url == True):                        
                    tokenObject.values.loc[index].append('URL')
                    tokenObject.types.loc[index].append('URL')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.like_email == True):                      
                    tokenObject.values.loc[index].append('EMAIL')
                    tokenObject.types.loc[index].append('EMAIL')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.like_num == True):                     
                    tokenObject.values.loc[index].append('NUM')
                    tokenObject.types.loc[index].append('NUM')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'PROPN' or t.pos_ == 'GPE' or t.pos_ == 'ORG'):  
                    tokenObject.values.loc[index].append('NAME-ENT')
                    tokenObject.types.loc[index].append('NAME-ENT')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'NOUN'):                 
                    tokenObject.values.loc[index].append(t.lemma_)
                    tokenObject.types.loc[index].append('NOUN')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'VERB'):                 
                    tokenObject.values.loc[index].append(t.lemma_)
                    tokenObject.types.loc[index].append('VERB')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'ADJ'):                 
                    tokenObject.values.loc[index].append(t.text)
                    tokenObject.types.loc[index].append('ADJ')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'ADV'):                 
                    tokenObject.values.loc[index].append(t.lemma_)
                    tokenObject.types.loc[index].append('ADV')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'PRON'):                 
                    tokenObject.values.loc[index].append('PRON')
                    tokenObject.types.loc[index].append('PRON')
                    tokenObject.tokens.loc[index].append(t)
                elif(t.pos_ == 'PUNCT'):
                    continue
                elif(t.pos_ == 'INTJ'):                 
                    tokenObject.values.loc[index].append(t.lemma_)
                    tokenObject.types.loc[index].append('INTJ')
                    tokenObject.tokens.loc[index].append(t)
                else:
                    tokenObject.values.loc[index].append(t.lemma_)
                    tokenObject.types.loc[index].append('X')
                    tokenObject.tokens.loc[index].append(t)
    return tokenObject
````

````python
x = Lemma(data)
x.values.to_csv('Values.csv')
x.types.to_csv('Types.csv')
x.tokens.to_csv('Tokens.csv')
````

## Part 2: Bag of words, Tf-idf, KNN

* This is the function to calculate the sentence vector from tokens of each tweet. After tokenize the doc, spacy give us back the token typpe, it has a vector attribute represented the value of this word in vector space. Then we calculate the vector of the whole sentence, here I use the average of word-vector method

````python
def cal_sentence_vectors(tokenObject):
    '''
    Calculate sentence vector using word vector of Spacy
    '''
    tokens = tokenObject.tokens.copy()
    l = len(tokens)
    sentence_vectors = pd.Series(data=([] for index in range(0,l)))
    for index in range(0,l):
        temp = tokens.loc[index]
        if(len(temp) != 0):
            l2 = len(temp)
            sentence_vectors[index] = temp[0].vector
            for index2 in range(1, l2):
                sentence_vectors[index] += temp[index2].vector
            sentence_vectors[index] /= l2
        else:
            sentence_vectors[index] = np.zeros(384)
            
    return sentence_vectors 

    
sentence_vectors_spacy = cal_sentence_vectors(x)
````

def compute_IDF(list_sentence_bow):
    idf_dict = {}
    # N: number of sentences/docs
    N = len(list_sentence_bow)
    
    #count number of documents that contain this word
    idf_dict = dict.fromkeys(list_sentence_bow[0].keys(), 0)
    for sentence in list_sentence_bow:
        for word, count in sentence.items():
            if count > 0:
                idf_dict[word] += 1

        idf_dict[word] = math.log(N/float(count))

## Bag of words, Tf-idf, KNN using sklearn

* The corpus cointain all the distinct word in our data, we get it from the create_dict() function above, ten we will create 2 models, bag of word and tf-idf and fit this model with the corpus

````python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [' '.join(create_dict(x))]
bag_of_words_model = CountVectorizer()
tf_idf_model = TfidfVectorizer()
bag_of_words_model.fit(corpus)
tf_idf_model.fit(corpus)
````

* This function turn the list of tweet into list of vector, the input value is the model (Bow or Tfidf) and the values in the Token object that we create. The output is the list of vectors.

````python
def cal_list_vectors(model, token_list):
    temp = token_list.copy()
    l = len(temp) 
    result = []
    for index in range(0, l):
        t = [' '.join(temp.loc[index])]
        result.append(model.transform(t).toarray())
    
    return result

list_bow = cal_list_vectors(bag_of_words_model, x.values)
list_tfidf = cal_list_vectors(tf_idf_model, x.values)
````

* We must turn the list into numpy array so that it can run with the sklearn, and also we need to reshape the Bow and tfidf vector from 3d to 2d
